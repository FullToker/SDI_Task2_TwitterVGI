```python
import pandas as pd
import missingno as msno
```

```python

df = pd.read_csv('../Load_Pre/data/geo_UEFA_tweets.csv')
# Filter for coordinates in the United States
# US boundaries: approximately latitude 24-49, longitude -125 to -66
def is_in_us(coord_str):
    try:
        if pd.isna(coord_str):
            return False
        lat, lon = map(float, coord_str.split(','))
        return 24 <= lat <= 49 and -125 <= lon <= -66
    except:
        return False

us_df = df[df['coordinates'].apply(is_in_us)]
us_df
```
